```python
import pandas as pd
import numpy as np
import difflib
```

```python
amenities = pd.read_json("food-amenities-vancouver.json", lines=True)
amenities
```

```python
canadian_chains = pd.read_csv("CanadaFoodChains.csv")
canadian_chains
```

```python
#filter canadian food chains
amenities['closest_name'] = amenities['name'].apply(lambda title: difflib.get_close_matches(title, canadian_chains['foodChains'].values, 1, .85))
amenities = amenities.sort_values(by=['closest_name'], ascending=False)
amenities['closest_name'] = amenities['closest_name'].apply(lambda title: ''.join(title)) 
canadian_chains = amenities[amenities['closest_name'] != '']
canadian_chains.to_json('chain-resturants-vancouver.json', orient='records', lines=True)
canadian_chains
```
