```python
import pandas as pd
import numpy as np
```

```python
amenities = pd.read_json("amenities-vancouver.json.gz", lines=True)
```

```python
travel_list = ['bus_station', 'car_rental', 'bicycle_rental', 'car_sharing']
travel_df = amenities[amenities['amenity'].isin(travel_list)]
travel_df = travel_df.drop(columns=['timestamp'])
travel_df
```

```python
travel_df = travel_df.sort_values(by=['name'])
travel_df.to_json('transportation-amenities-vancouver.json', orient='records', lines=True)
```
