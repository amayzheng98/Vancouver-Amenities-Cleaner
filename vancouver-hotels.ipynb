```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# access all hotel locations
accomodations = pd.read_csv('HLLBC_ACC.csv')
accomodations = accomodations.filter(items=['NAME','DESCRIPTN','LOCALITY','ADDRESS','LATITUDE','LONGITUDE'])
accomodations['DESCRIPTN'].unique()
```

```python
# filter to all the hotels, resorts, and bed & breakfasts
hotel_list = [
    'Bed & Breakfasts; Campgrounds & RV Parks - Private;', 
    'Bed & Breakfasts;', 'Hotels & Motels;','Hostels;', 
    'Hotels & Motels; Resorts;','Hotels & Motels; Cabins & Cottages;',
    'Hotels & Motels; Cabins & Cottages; Campgrounds & RV Parks - Private;',
    'Hotels & Motels; Campgrounds & RV Parks - Private;',
    'Resorts; Vacation Rentals; Cabins & Cottages; Campgrounds & RV Parks - Private;',
    'Resorts;','Hotels & Motels; Hostels;','Hotels & Motels; Resorts; Vacation Rentals;',
    'Hotels & Motels; Resorts; Cabins & Cottages; Campgrounds & RV Parks - Private;',
    'Hotels & Motels; Vacation Rentals; Cabins & Cottages;',
    'Resorts; Campgrounds & RV Parks - Private;',
    'Resorts; Cabins & Cottages; Campgrounds & RV Parks - Private;',
    'Hotels & Motels; Resorts; Cabins & Cottages;',
    'Resorts; Vacation Rentals; Accommodation Management Companies;',
    'Resorts; Vacation Rentals;',
    'Hotels & Motels; Vacation Rentals;',
    'Bed & Breakfasts; Vacation Rentals;',
    'Bed & Breakfasts; Vacation Rentals; Cabins & Cottages;',
    'Hostels; Vacation Rentals;',
    'Bed & Breakfasts; Cabins & Cottages; Campgrounds & RV Parks - Private;',
    'Bed & Breakfasts; Hotels & Motels;'
]
```

```python
# filter all greater Vancouver cities
greater_vancouver_list = [
    'Burnaby',
    'Bowen Island',
    'Coquitlam',
    'Delta',
    'Langley',
    'Maple Ridge',
    'New Westminister',
    'North Vancouver',
    'Pitt Meadows',
    'Port Moody',
    'Richmond',
    'Surrey',
    'Vancouver',
    'West Vancouver',
    'White Rock'
,]
```

```python
hotel_df = accomodations[accomodations['DESCRIPTN'].isin(hotel_list)]
hotel_df = accomodations[accomodations['LOCALITY'].isin(greater_vancouver_list)]
hotel_df = hotel_df.rename(columns={'LATITUDE': 'lat', 'LONGITUDE': 'lon','NAME':'name','DESCRIPTN':'amenity','ADDRESS':'tags'})
hotel_df
```

```python
hotel_df.to_json('vancouver-hotels.json', orient='records', lines=True)
```
